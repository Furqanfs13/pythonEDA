```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/banking_data.csv')
print(df.head())
```

What is the distribution of age among the clients?


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of Age Among Clients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

How does the job type vary among the clients?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['job'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of Age Among Clients')
plt.xlabel('jobs')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
freq = df['job'].value_counts()
print(freq)
plt.show()
```

What is the marital status distribution of the clients?

```python
plt.figure(figsize=(8, 8))
marital_status_counts = df['marital'].value_counts()
plt.pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Marital Status Distribution of Clients')
plt.show()
status = df['marital'].value_counts()
print(status)
```

What is the level of education among the clients?

```python
plt.figure(figsize=(8, 8))
education_level_counts = df['education'].value_counts()
plt.pie(education_level_counts, labels=education_level_counts.index, autopct='%1.1f%%', colors=['skyblue','red', 'lightgreen', 'lightcoral'])
plt.title('Education level of Clients')
plt.show()
lvl = df['education'].value_counts()
print(lvl)
```

What proportion of clients have credit in default?

```python
prop = df['default'].value_counts()
print(prop)
plt.figure(figsize=(8, 8))
plt.pie(prop, labels=prop.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Proportion of Clients with Credit in Default')
plt.show()
```

What is the distribution of average yearly balance among the clients?

```python

plt.figure(figsize=(30, 60))
sns.histplot(df['balance'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of Average Yearly Balance Among Clients')
plt.xlabel('Average Yearly Balance')
plt.ylabel('Frequency')


min_balance = min(df['balance'])
max_balance = max(df['balance'])
plt.xlim(min_balance, max_balance)
plt.ylim(0, 7000)
plt.xticks(np.arange(min_balance, max_balance + 2000, 2000), rotation=45, ha='right')
plt.yticks(np.arange(0, 8000 + 750, 750), rotation=45, ha='right')
plt.grid(True)
plt.show()
```

What is the distribution of the last contact day of the month?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['day'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of last contact day of the month')
plt.xlabel('day')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
freq = df['day'].value_counts()
print(freq)
plt.show()
```

How does the last contact month vary among the clients?

```python
count = df.groupby('month')['month'].count()
print(count)

plt.figure(figsize=(10, 6))
sns.histplot(df['month'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of last contact month')
plt.xlabel('month')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(True)

```

What is the distribution of the duration of the last contact?

```python
count = df.groupby('duration')['duration'].count()
print(count)

plt.figure(figsize=(10, 6))
sns.histplot(df['duration'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of the duration of last contact')
plt.xlabel('duration')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(True)


```

How many contacts were performed during the campaign for each client?

What is the distribution of the number of days passed since the client was last contacted from a previous campaign?

```python
count = df.groupby('pdays')['pdays'].count()
print(count)

plt.figure(figsize=(10, 6))
sns.histplot(df['pdays'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of the duration of last contact')
plt.xlabel('pdays')
plt.ylabel('Frequency')
mini = df['pdays'].min()
maxi = df['pdays'].max()

plt.xticks(np.arange(mini,maxi + 1000,1000), rotation=45, ha='right')
plt.grid(True)

```
